## Структура среды (env.py)

Вся логика вынесена в `env.py` — его импортируют все 4 ноутбука.

| Класс / объект | Что делает |
|---|---|
| `DEFAULT_MAP` | 10×10 numpy-массив: 0=проход, 1=стена |
| `Tank(row, col, angle)` | Позиция и угол поворота танка |
| `Goal(row, col)` | Позиция цели |
| `TankEnv` | Среда: `reset()`, `step(action)`, `render()` |

**Действия:** 0=вперёд, 1=назад, 2=влево, 3=вправо  
**Наблюдение:** `(row, col, angle_idx, goal_row, goal_col)`

```python
env = TankEnv(tank_start=(1,1), goal_pos=(8,8))
obs = env.reset()

print('Начальное состояние:', obs)
print(f'Танк: row={env.tank.row}, col={env.tank.col}, angle={env.tank.angle}')
print(f'Цель: row={env.goal.row}, col={env.goal.col}')

frame = env.render()
plt.figure(figsize=(5,5))
plt.imshow(frame)
plt.axis('off')
plt.title('Начальное состояние')
plt.tight_layout()
plt.show()
```
